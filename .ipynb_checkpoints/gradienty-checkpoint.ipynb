# Utworzenie gradientów przy użyciu własnej klasy i plt.imshow


```python
from skimage import io
from skimage import color
from HSVANDRGB import *
from matplotlib import pyplot as plt
import numpy as np
shapes=(70,650,3)#wymiary gradientów
```

Gradient Pierwszy

```python
CzaBia=laduj_opcje("0-1",[650])
CzaBia=kreacja3x([CzaBia,CzaBia,CzaBia],shapes)
plt.imshow(CzaBia)
```

Gradient Drugi

```python
re=laduj_opcje("0, 0-1",[650//2,650//2]) 
gr=laduj_opcje("1-0, 0", [650//2,650//2])
bl=laduj_opcje("0-1, 1-0",[650//2,650//2])
regebl=kreacja3x([re,gr,bl],shapes)
plt.imshow(regebl)
```

Gradient Trzeci

```python
re=laduj_opcje("0, 0-1, 0+x",[650//2,650//4,(650//4+1,1)])
gr=laduj_opcje("0+x, 1-0, 0",[(650//4+1,1),650//4,650//2])
bl=laduj_opcje("0-1, 0+x, 1-0",[650//4+1,(650//2,1),650//4])
regebl=kreacja3x([re,gr,bl],shapes)
plt.imshow(regebl)
```

Gradient Czwarty (taki sam jak na obrazku) użycie funkcji my_append

```python
re=laduj_opcje("0+x, 1-0, 0, 0-1, 0+x, 1-0",[(93,1),93,186,93,(93,1),92])
gr=laduj_opcje("1-0, 0, 0-1, 0+x, 1-0, 0",[93,93,93,(186,1),93,92])
bl=laduj_opcje("0+x, 1-0, 0", [(93*3,1),93,93*3-1])
regebl=kreacja3x([re,gr,bl],shapes)
plt.imshow(regebl)
```

```python
hue=laduj_opcje("0",[650])
sat=laduj_opcje("0",[650])
val=laduj_opcje("0-1",[650])
regebl=HSV(kreacja3x([hue,sat,val],shapes)).toRGB()
plt.imshow(regebl)
```

```python
hue=laduj_opcje("x-x",[(120,360,650)])
sat=laduj_opcje("0+x",[(650,1)])
val=laduj_opcje("0+x",[(650,1)])
regebl=HSV(kreacja3x([hue,sat,val],shapes)).toRGB()
plt.imshow(regebl)
```

```python
hue=laduj_opcje("x-x",[(120,0,650)])
sat=laduj_opcje("0+x",[(650,0.502)])
val=laduj_opcje("0+x",[(650,1)])
regebl=HSV(kreacja3x([hue,sat,val],shapes)).toRGB()
plt.imshow(regebl)
```

```python
hue=laduj_opcje("x-x",[(0,315,650)])
sat=laduj_opcje("x-x",[(1,0.016,650)])
val=laduj_opcje("0+x",[(650,1)])
regebl=HSV(kreacja3x([hue,sat,val],shapes)).toRGB()
plt.imshow(regebl)
```

Gradient Wysokości

```python
hue=laduj_opcje("x-x",[(120,0,650)])
sat=laduj_opcje("x-x, 0+x",[(0.635,1,(650*2)//6),(650-(650*2)//6,1)])
val=laduj_opcje("x-x",[(0.615,1,650)])
regebl=HSV(kreacja3x([hue,sat,val],shapes)).toRGB()
plt.imshow(regebl)
```

```python
a1=Dem()
a=a1.punkt_pierwszy()
hue=interpreter_warunkowy(a,"x-x",[(120,0,650)])
sat=interpreter_warunkowy(a,"0+x",[(650,1)])
val=interpreter_warunkowy(a,"0+x", [(650,1)])

        
regebl=HSV(kreacja3x([hue,sat,val],(500,500,3))).toRGB()

plt.imshow(regebl)
```

```python
a2=a1.tablica.copy()
a2[:,1:]=a2[:,:-1]
a3=kreacja3x([hue,sat,val],(500,500,3))
a3[:,:,2]=np.minimum(procent(a1.tablica-a2)*2,1.0)
plt.imshow(HSV(a3).toRGB())
```

```python
a4=a1.tablica.copy()
a4[:-1,:]=a4[1:,:]
print(np.max(np.arctan2(a4-a1.tablica,a1.o/100))/(2*np.pi)*360)
sat=np.arctan2(a4-a1.tablica,a1.o/100)/(2*np.pi)*360
val=np.minimum(procent(a1.tablica-a2)*2,1.0)
#za późno na punkt trzeci wysyłam, szkoda ocełniłbym siebie 1,66 punktu na ostatni potrzeba 3h
mat=(1-((sat>-6)*(sat<=84)*((sat+6)/90)))*(val)*(sat>-6) + (sat<=-6)
a4=kreacja3x([hue,mat,val],(500,500,3))
plt.imshow(HSV(a4).toRGB())
```
