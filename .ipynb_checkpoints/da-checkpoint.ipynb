### Kolory, obrazy

```python
def gen_img(n=9, start=-1.5, end=1.5, size=2000):
    # Inicjalizacja danych
    x = np.linspace(start, end, size)
    y = np.linspace(start, end, size)
    xx, yy = np.meshgrid(x, y)
    Z = xx+yy*1j # Tworzymy macierz wartości zespolonych
    
    for i in range(50):
        #Z =  ((n-1)/n)*Z + (1/n)*(1/(Z**(n-1))) # alternatywny obraz
        Z = ((n**2-12/n**2)/n**2) * Z - 3/n**2 + (1/n**3)*(1/(Z**(n-1)))

    # Argument liczby zespolonej
    img = np.angle(Z)
    # Normalizacja do zakresu 0-255
    return 255*(img - np.min(img))/np.ptp(img)
```

```python
# Można również przetestować funkcję dla innych n niż domyślne
img = gen_img()
```

```python
# Funkcja zwraca dwuwymiarową tablicę wartości z zakresu 0-255
print(img)
```

```python
# Wyświetlanie numpy array jako obrazu
plt.imshow(img)
```

```python
# Zmiana mapy kolorów
plt.imshow(img, cmap="terrain")
```

8. Korzystając z numpy wygeneruj poniższy obraz

![title](goal.png)

```python
star_trek=np.zeros((100,100,3),dtype=float)
czerwony=np.array([[(1.0 if (i-50)**2+(j-50)**2<=20**2 else 0.0)for j in range(100)] for i in range(100)])
zielony=np.array([[(1.0 if (i-50)**2+(j-50)**2<=35**2 else 0.0)for j in range(100)] for i in range(100)])
niebieski=np.array([[(1.0 if (i-50)**2+(j-50)**2<=50**2 else 0.0)for j in range(100)] for i in range(100)])
star_trek[:,:,0]=czerwony
star_trek[:,:,2]=niebieski-zielony*niebieski
star_trek[:,:,1]=zielony-czerwony*zielony
plt.imshow(star_trek)
```

```python
samp,leto= wavfile.read("leto.wav")
playsound("leto.wav")
```

5. Zastosuj efekt fade in (stopniowe zgłaśnianie) przemnażając próbki przez odpowiednie wagi i zapisując efekt do pliku .wav. Pamiętaj że nagranie ma dwie ścieżki.

```python
glowna=np.zeros(leto.shape)
glowna[:,0]=np.linspace(0,100,leto.shape[0])*leto[:,0]
glowna[:,1]=leto[:,1]*np.linspace(0,100,leto.shape[0])
wavfile.write("test7.wav", samp, glowna.astype('int16'))
playsound("test7.wav")
```

6. Powtórz poprzednie polecenie, tym razem stosując fade-in logarytmiczny. Jaka jest różnica w porównaniu do poprzedniego podejścia? Dlaczego tak się dzieje? 

```python
glowna1=np.zeros(leto.shape)
glowna1[:,0]=np.log10(np.linspace(0,100,leto.shape[0]))*leto[:,0]
glowna1[:,1]=np.log10(np.linspace(0,100,leto.shape[0]))*leto[:,1]
wavfile.write("test12.wav", samp, glowna1.astype('int16'))
playsound("test12.wav")
```

7. Odejmij od siebie wartości próbek w obu kanałach, podziel wynik przez dwa i zapisz do pliku .wav. Pamiętaj żeby przed zapisem do pliku przekonwertować wynik tej operacji na <b>int</b> korzystając z <b>astype('int16')</b>. Co się stało z nagraniem? Jaka jest tego przyczyna?

```python
wavfile.write("test6.wav", samp, ((glowna-glowna1)/2).astype('int16'))
playsound("test6.wav")
```
